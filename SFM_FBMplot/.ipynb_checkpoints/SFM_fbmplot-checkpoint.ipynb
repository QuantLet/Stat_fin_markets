```python
from fbm import FBM

H=0.001
n=10000
f = FBM(n=n, hurst=H, length=1, method='daviesharte')
# or
f = FBM(n, H)

# Generate a fBm realization
fbm_sample = f.fbm()

# Generate a fGn realization
fgn_sample = f.fgn()

# Get the times associated with the fBm
t_values = f.times()
```

```python
import matplotlib.pyplot as plt 
# This line is necessary for the plot to appear in a Jupyter notebook
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline

pylab.rcParams['figure.figsize'] = (12, 6)   # Change the size of plots
# Plot
f, ax = plt.subplots()
plt.title('Simulated price H=%5.3f'%(H))
ax.plot(t_values, fbm_sample, color="blue")

ax.grid(True)
plt.show()
```

```python

f, ax = plt.subplots()
plt.title('Simulated return H=%5.3f'%(H))
ax.plot(t_values[1:], fgn_sample, color="blue")

ax.grid(True)
plt.show()
```
