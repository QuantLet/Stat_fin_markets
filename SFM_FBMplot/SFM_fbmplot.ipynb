```python
! pip install fbm
from fbm import FBM

H=0.5
n=1000
#f = FBM(n=n, hurst=H, length=1, method='daviesharte')
# or
f = FBM(n, H)

# Generate a fBm realization
fbm_sample = f.fbm()

# Generate a fGn realization
fgn_sample = f.fgn()

# Get the times associated with the fBm
t_values = f.times()
```


### Fractional Brownian Motion (fBm)

Fractional Brownian Motion $ B_H(t) $ is defined as a continuous-time Gaussian process on $[0, T]$ which starts at 0, has expectation 0 for all $ t $, and follows the covariance function:

$
E[B_H(t) B_H(s)] = \frac{1}{2} \left( |t|^{2H} + |s|^{2H} - |t-s|^{2H} \right)
$

Where:
- $ E $ denotes the expected value.
- $ t $ and $ s $ are times within the interval $[0, T]$.
- $ H $ is the Hurst parameter, which indicates the degree of memory in the process.

### Fractional Gaussian Noise (fGn)

Fractional Gaussian Noise $ X_k $ is the discrete increment process derived from fBm, defined by:

$
X_k = B_H(k+1) - B_H(k)
$

Where:
- $B_H(k)$ is the fractional Brownian motion at discrete time $k$.
- $X_k $ represents the difference between two successive values of the fractional Brownian motion, thus characterizing the increments.

These mathematical definitions are crucial for simulating and analyzing data with long memory properties, making them applicable in diverse fields such as finance, telecommunications, and environmental science. These processes help in modeling phenomena where past events influence future outcomes more significantly than in standard Brownian motion, where there is no memory effect.
