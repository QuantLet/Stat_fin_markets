```python

from scipy.stats import genpareto


import numpy as np
import matplotlib.pyplot as plt

csi=0.3
beta=0

fig, ax = plt.subplots(figsize=(20, 10))

x = np.linspace(genpareto.ppf(0.01, csi,beta),
                genpareto.ppf(0.99,csi,beta), 1000)
ax.plot(x, genpareto.pdf(x,csi,beta),
       'r-', lw=5, alpha=0.8, label='Pareto pdf')
ax.legend(loc='best', frameon=False)



csi=0
x = np.linspace(genpareto.ppf(0.01, csi,beta),
                genpareto.ppf(0.99,csi,beta), 1000)
ax.plot(x, genpareto.pdf(x,csi,beta),
       'g-', lw=5, alpha=0.8, label='Exponential pdf')
ax.legend(loc='best', frameon=False)


csi=-0.3
x = np.linspace(genpareto.ppf(0.01, csi,beta),
                genpareto.ppf(0.99,csi,beta), 1000)
ax.plot(x, genpareto.pdf(x,csi,beta),
       'b-', lw=5, alpha=0.8, label='Pareto II pdf')
ax.legend(loc='best', frameon=False)



plt.show()
```
